```
# created two arrays holding information on GPU's price and # of GFLOPs
gpu_price = [79, 109, 139, 299, 379, 499, 599, 699, 1200, 1200, 2999, 599, 799, 1199]

# The GPUs GFLOPS are taken as FP32 single value operations from link:
# https://en.wikipedia.org/wiki/List_of_Nvidia_graphics_processing_units
gpu_GFLOPS = [38, 1862, 2138, 4375, 6463, 8186, 8873, 11340, 10974, 11366, 12288, 14930, 20137, 26895]

# List of GPUs as the reference to print out after finding the combination of GPUs to use

# pos: index of current item
# selected: list of indices of selected items
MAX_PRICE = int(input('What is your budget (to the nearest dollar)? : $'))

#After solving the 0/1 problem you should start thinking about being able to take the Bigger knapscack problem into consideration
# create at least 10 test cases to see if code is 100% correct
# look into doing a bottom-up approach to this program (right now its recursive)

def select_processor(pos, selected):
    
    # calculate totals
    total_GFLOPS = 0
    total_price = 0
    for i in selected:
        total_price += gpu_price[i] # make these into parameters so test cases can be seen w/ different arrangements
        total_GFLOPS += gpu_GFLOPS[i] 
    
    # base case: no items left to analyze
    if total_price > MAX_PRICE:
        return (0,0)
    
    if pos >= len(gpu_GFLOPS):
      return (total_price, total_GFLOPS, selected)
      
    # recursive case
    ans1 = select_processor(pos+1, selected+[pos]) # switch on
    
    # .copy() creates a copy of the list constructed so far as to not
    # mess any of the values up / making no inconsistencies later on
    ans2 = select_processor(pos+1, selected.copy()) # switch off
    
    # need to find the ans with greater value
    
    if ans1[0] > ans2[0]:
        return ans1
    else: 
        return ans2

# This gives the values of the max GPUs user can get along with # of GFLOPs
ans = select_processor(0, [])
print(ans)
    
    
#create another array on top of other arrays, creates sort of 'cost' 
#create an array with that function / multiple cases
#check the cost array w/ indexes
#think about it like working backwards as a possible solution
```

```
#Create the arrays that will be used 

# Using general dollar cost from NVIDIA website
cost = [87, 109, 139, 299, 379, 499, 599, 699, 1200, 1200, 2999, 599, 799, 1199]

# Numbers corresponding to the list from GPU comparison table
gflops = [38, 1862, 2138, 4375, 6463, 8186, 8873, 11340, 10974, 11366, 12288, 
          14930, 20137, 26895]

# Bandwidth of each gpu from comparison table
gpu_bw = [23, 84, 112, 216, 256, 256, 320, 484, 480, 547, 653, 448, 448, 616]
gpu_model = ["GTX 1030","GTX 1050","GTX 1050TI","GTX 1060","GTX 1070","GTX 1070TI",
             "GTX 1080","GTX 1080TI","TITAN X","TITAN XP","TITAN V","RTX 2070","RTX 2080", "RTX 2080TI"]



#num_items = the amount of gpus being compared
num_items  = len(gflops)


#should return the max value of glfops that can be acheived by budget
def gpu_combo(budget):
    
    # gpu_glfop[i] should store the maximum
    # value with knapsack capacity i.
    
    #base case
    gpu_gflop = [{"gflops": 0, 
                  "selected": [0 for i in range(num_items)]}]
    
    #gpu_gflop = [(0, [0 for i in [range(len(gflops))] for i in range(budget + 1)]
    total_bw = 0
    # Fill gpu_gflop[] using a recursive formula
    # Goes through the dollar budget amount just like normal for loop
    for i in range(budget + 1):
      # Appends the new gpu that will be selected to the array
      gpu_gflop.append(gpu_gflop[-1])
      # Goes through the number of different GPUs in array
      for j in range(num_items):
           if (cost[j] <= i) and ((gpu_gflop[i - cost[j]]["gflops"] + gflops[j]) > gpu_gflop[i]["gflops"]):
               #gpu_gflop[i] = (max(gpu_gflop[i], gpu_gflop[i - cost[j]] + gflops[j]))
                new_selected = gpu_gflop[i - cost[j]]["selected"][:] # [:] Makes a copy of the list
                new_selected[j] += 1
                new_gflops = gpu_gflop[i - cost[j]]["gflops"]
                new_gflops += gflops[j]
                gpu_gflop[i] = {"gflops": new_gflops, "selected": new_selected}
                
    
    return gpu_gflop[budget]   
  
budget = int(input('What is your budget (to the nearest dollar)? : $'))

###COPY "SELECTED" FROM ANS INTO ANOTHER VAR AND EVALUATE FROM THERE###
ans = gpu_combo(budget)
listed = ans["selected"][:]

total_bw = 0
for i in range(num_items):
    if listed[i] > 0:
        total_bw += (gpu_bw[i] * listed[i])
        gpu_select = gpu_model[i]
        print(gpu_select, 'x', listed[i])
        
print(total_bw, 'GB/s')


print(ans)




```
